```python
import pandas as pd
```

```python
filepath = './Resources/Border_Crossing_Entry_Data.csv'

# Import Data
raw_df = pd.read_csv(filepath)

raw_df.head()
```

```python
raw_df.info()
```

```python
# Split date column into month and year
raw_df[['month','year']] = raw_df.Date.str.split(' ', expand=True)

raw_df.head()
```

```python
drop_df = raw_df.drop(columns='Date')
```

```python
# Turn year into an integer
drop_df['year'] = drop_df['year'].astype(int)
```

```python
drop_df.info()
```

```python
filter = drop_df.year > 2017

filter_df = drop_df[filter]
filter_df.head()
```

```python
filter_df.columns = [col.lower().replace(' ','_') for col in filter_df.columns]
```

```python
filter_df.to_csv('cleaned_border_data.csv', index=False)
```
